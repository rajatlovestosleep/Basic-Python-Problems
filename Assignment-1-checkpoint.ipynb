## Question 2

```python
first=input("Enter first string: ")
second=input("Enter second string: ")
new_string=first+" "+second
print("Concatenated String:",new_string)
print("Length of new string:",len(new_string))
print("Uppercase:",new_string.upper())
print("Contains 'Albus':","Albus" in new_string)
```

### Description  
- **Task:** Take two strings as input from the user, concatenate them with a space, and store the result in a new variable called `new_string`. Then:  
  1. Print the value of `new_string`.  
  2. Print the length of `new_string`.  
  3. Convert `new_string` to uppercase.  
  4. Check whether `"Albus"` is present in `new_string`.  
- **Methods Used:**  
  - `input()` → to take user input for creating and concatenating the strings.  
  - `+` operator → to concatenate the two strings with a space between them.  
  - `len()` → to find the length of the new concatenated string.  
  - `.upper()` → to convert the new concatenated string into uppercase.  
  - `'in'` → to check for substring presence in `new_string`.  
- **Syntax:**  
  - `variable = input("message")` → takes user input.
  - `string1 + " " + string2` → concatenates the two strings with a space between them. 
  - `len(string)` → finds the length of the new concatenated string. 
  - `string.upper()` → converts the new concatenated string into uppercase. 
  - `"substring" in string` → to check for substring presence in `new_string`.
- **What the code has done:**  
  The program takes two user inputs, joins them with a space between, stores them in `new_string`, prints its value, prints its length, converts it to uppercase, and checks if `"Albus"` is present.  
- **Input Example:**  
  `"Walter"`, `"White"`  
- **Output Example:**  
  - Concatenated string: `"Walter White"`  
  - Length: `12`  
  - Uppercase: `"WALTER WHITE"`  
  - Substring check (`'Albus' in new_string`): `False`

## Question 2

```python
numbers=[1, 2, 3, 4, 5, 6]
half=len(numbers)//2
swapped=numbers[half:]+numbers[:half]
print("List after cyclic swap:",swapped)
```

### Description  
- **Task:** Cyclically swap the first half of a list with the second half.  
- **Method Used:**  
  - `len()` → to calculate the midpoint of the list.
  - `//` operator → to get the midpoint in integer in case of odd elements in list.
  - List slicing → to split the list into two halves.  
  - `+` operator → to join the second half first, then the first half.
- **Syntax:**  
  - `len(list)` → returns the number of elements in the list.  
  - `list[:n]` → selects the first n elements.  
  - `list[n:]` → selects the elements from index n to the end.  
  - `list2 + list1` → concatenates two lists.  
- **What the code has done:**  
  The program defines a list, calculates its halfway point using `len() // 2`, splits the list into two halves using slicing, and concatenates them in reversed order. The output is the list with its two halves swapped cyclically.  
- **Input Example:**  
  `[1, 2, 3, 4, 5, 6]`  
- **Output Example:**  
  `[4, 5, 6, 1, 2, 3]`

### Description

  - `list()` → to convert the range into a list.  
  - `sum()` → to calculate the total sum of the list elements.  
  - `len()` → to find the number of elements in the list.  
  - `/` → to calculate the average by dividing the sum by the number of elements.  
- **Syntax:**  
  - `numbers=list(range(start,end))` → creates a list of numbers from start to end-1.  
  - `total=sum()` → returns the total sum of list elements.  
  - `average = total / len(numbers)` → calculates the average.  
- **What the code has done:**  
  The program creates a list of numbers from 1 to 35, calculates the sum using the `sum()` method, finds the number of elements using `len()`, computes the average by dividing the sum by the length, and prints both sum and average with labels.  
- **Input Example:**  
  `[1, 2, 3, ..., 35]`  
- **Output Example:**
  `Sum:  630
Average:  18.0`
